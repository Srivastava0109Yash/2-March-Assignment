### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular data visualization library in Python that provides a variety of plotting functions to create different types of plots such as line plots, scatter plots, bar plots, histograms, and many more. It is widely used for creating visualizations in fields like data science, machine learning, and scientific research.
Matplotlib is used to create high-quality visualizations that can help to represent data in a clear and effective manner. Some of the common use cases for Matplotlib include exploring data, identifying trends and patterns, comparing data across different groups, and presenting findings to others.

___Here are five types of plots that can be plotted using the Pyplot module of Matplotlib___:

__Line Plot__ - A line plot is a type of plot that displays data points as markers connected by lines. It is useful for showing how a variable changes over time or across different groups.

__Scatter Plot__ - A scatter plot is a type of plot that displays data points as individual markers on a two-dimensional plane. It is useful for showing the relationship between two variables.

__Bar Plot__ - A bar plot is a type of plot that displays data as rectangular bars with lengths proportional to the values they represent. It is useful for comparing data across different groups.

__Histogram__ - A histogram is a type of plot that displays the distribution of a variable by dividing it into intervals and counting the number of data points that fall into each interval. It is useful for visualizing the shape of a distribution and identifying outliers.

__Pie Chart__ - A pie chart is a type of plot that displays data as slices of a circular pie. The size of each slice is proportional to the value it represents. It is useful for showing how a variable is divided into different categories.

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

A scatter plot is a type of data visualization that displays the relationship between two numerical variables. It is a graph that uses dots to represent data points on a two-dimensional coordinate system, where one variable is plotted on the x-axis and the other variable is plotted on the y-axis.

In a scatter plot, each data point represents an observation in the data set, and its position on the graph corresponds to the values of the two variables being plotted. The placement of the dots on the graph can reveal patterns or trends in the data, such as the presence of a positive or negative correlation between the two variables.It is useful for identifying the relationship between two variables, detecting outliers, and visualizing the spread and distribution of the data.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.title("Scatter Plot of X and Y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

__Bar plot__, also known as a bar chart, is a type of chart used to display data using rectangular bars. Each bar represents a category or group of data, and the height or length of the bar corresponds to the value or frequency of the data.

__Use:__
Bar plots are useful for comparing data between different categories or groups. They are often used to visualize data that can be divided into discrete categories, such as nominal or ordinal data. Bar plots are also useful for displaying data that has a natural order, such as time series data or ranked data,a simple and effective way to display categorical data and make comparisons between groups or categories.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
import matplotlib.pyplot as plt
```

#### Bar plot

```python
plt.bar(company,profit)
plt.title("Bar Plot of X and Y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

#### Horizontal bar plot.

```python
plt.barh(company,profit)
plt.title("Horizontal Bar Plot of X and Y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

__Box plot__, also known as a box-and-whisker plot, is a type of chart used to display the distribution of a dataset. It displays the five-number summary of a dataset (minimum, first quartile, median, third quartile, and maximum) as well as any outliers.

#### Use
Box plots are useful for identifying the spread and skewness of a dataset, as well as detecting outliers. They are particularly useful for comparing the distribution of a variable across different groups or categories.

box plots are a useful tool for visualizing the distribution of a dataset and comparing it across different groups or categories. They are particularly helpful when dealing with datasets with a large number of observations or when trying to identify outliers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data = [box1, box2]
fig, ax = plt.subplots()

ax.boxplot(data)
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')
ax.set_title('Box Plot Example')
plt.show()
```
